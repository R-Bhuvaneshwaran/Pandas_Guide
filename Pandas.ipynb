The **info()** method provides a concise summary of a DataFrame, including:

    Index type and range: Information about the DataFrame's index.
    Column names and their data types (dtypes): Identifies whether columns are numeric, object (strings), datetime, etc.
    Non-null counts: Shows the number of non-missing values in each column, which helps in identifying missing data.
    Memory usage: Estimates the memory consumed by the DataFrame.

```python
df.info()
```

The describe() method generates descriptive statistics of the DataFrame's columns. By default, it focuses on numerical columns, providing: 

    count: Number of non-null values.
    mean: Average value.
    std: Standard deviation.
    min: Minimum value.
    25% (Q1): 25th percentile (first quartile).
    50% (Q2): 50th percentile (median).
    75% (Q3): 75th percentile (third quartile).
    max: Maximum value.

## 🔁 8. Handling Time/Ordering (if present)  

## http://localhost:8888/notebooks/Desktop/Medium_Documents/Pandas_Time_Series.ipynb
    If dataset has dates, show datetime conversion:

```python
df['placed_date']=pd.date_range(start='2023-01-01', periods=len(df), freq='W')  #create dummy date
```

```python
df['Placement Status'].value_counts()
```

```python
filt_df=df[df['Placement Status']=="Not Placed"]
filt_df['placed_date']=pd.NaT
```

```python
df=df[df['Placement Status']=="Placed"]
df=pd.concat([df,filt_df],ignore_index=False)
```

```python
df['Placement Status'].value_counts()
```

```python
df.placed_date.value_counts()
```

```python
df
```

```python
df['placed_date'] = pd.to_datetime(df['placed_date'])
df.set_index('placed_date', inplace=True)
df.resample('M')['Salary'].mean()
```

## Relationship Between CGPA and Salary

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df, x='Sem1 GPA', y='Salary')
plt.title('CGPA vs Salary')
plt.xlabel('CGPA')
plt.ylabel('Salary')
plt.show()
```

```python
# Correlation
correlation = df[['Sem1 GPA', 'Salary']].corr()
correlation
```

## Salary Comparison by Gender

```python
sns.boxplot(data=df, x='Gender', y='Salary')
plt.title('Salary Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()
```
